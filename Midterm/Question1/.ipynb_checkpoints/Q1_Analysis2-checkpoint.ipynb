## After finding out the time duration and who sends maximum number of emails to whom, We will find out who sends and receives maximum emails amongst all the employees.

```python
import os
from email.parser import Parser
import email.utils
import time
import datetime
from dateutil.parser import parse
```

```python
import os
os.chdir("D:")
a=os.getcwd()
path= a+"enron_mail_20150507/maildir"
print(path)
#os.listdir(path)
```

```python
sent_mails = {}

received_mails= {}

### We will consider only the 'sent' items from the data to filter out all the spam and unsent items from the data####


for filename in os.listdir(path):
    for a in os.listdir(path+'/'+filename):
        if (a=='sent'):
                ab_path=path+'/'+filename+'/'+a
                count_sent = len([name for name in os.listdir(ab_path) if os.path.isfile(os.path.join(ab_path, name))])
                if filename not in sent_mails:  
                        sent_mails[filename] = count_sent
                else: sent_mails[filename]+=count_sent
        
        elif (a=='inbox')or(a=='notes_inbox'):
            ab_path=path+'/'+filename+'/'+a
            count_inbox = len([name for name in os.listdir(ab_path) if os.path.isfile(os.path.join(ab_path, name))])
            if filename not in received_mails:  
                received_mails[filename] = count_inbox
            else: received_mails[filename]+=count_inbox
```

```python
import matplotlib.pyplot as plt

b1 = [u[1] for u in sorted(sent_mails.items(), key=lambda x:x[1],reverse=True)[:10]]
b2 = [u[0] for u in sorted(sent_mails.items(), key=lambda x:x[1],reverse=True)[:10]]

#####################  TOP 10 USERS   ###########################
top10_sent = [u[0] for u in sorted(sent_mails.items(), key=lambda x:x[1],reverse=True)[:50]]


from pylab import rcParams
rcParams['figure.figsize'] = 16, 6

figure = plt.figure()
figure.suptitle('Top 10 Employees Who Sent Most Number of Emails', fontsize=20)
plt.xlabel('Employees', fontsize=16)
plt.ylabel('Number of Sent Mails', fontsize=16)
plt.bar(range(len(b1)), b1, align='center')
plt.xticks(range(len(b2)), b2,rotation='vertical')

plt.show()
```

```python
import matplotlib.pyplot as plt

c1 = [u[1] for u in sorted(received_mails.items(), key=lambda x:x[1],reverse=True)[:10]]
c2 = [u[0] for u in sorted(received_mails.items(), key=lambda x:x[1],reverse=True)[:10]]

################## TOP 10 USERS   ########################################
top10_received= [u[0] for u in sorted(received_mails.items(), key=lambda x:x[1],reverse=True)[:50]]


from pylab import rcParams
rcParams['figure.figsize'] = 16, 6

fig = plt.figure()
fig.suptitle('Top 10 Employees who received Most Number Of EMails', fontsize=20)
plt.xlabel('Employees', fontsize=16)
plt.ylabel('Number of Emails', fontsize=16)
plt.bar(range(len(c1)), c1, align='center')
plt.xticks(range(len(c2)), c2,rotation='vertical')

plt.show()
```

```python
intersection_list = set(top10_sent).intersection(top10_received)
print('The users with highest sent and received mail activity are:',intersection_list)
#thefile = open('C:\DataAnalysis4Python\kanakia_dhruv_spring2017\DataAnalysis4Python_Spring17\\HighActivity_Users.txt', 'w')

#for item in thefile:
    #thefile.write("%s\n" % item)
```
