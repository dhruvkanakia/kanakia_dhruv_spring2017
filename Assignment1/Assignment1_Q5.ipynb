```python
from nltk.corpus import stopwords

# Getting the English stop words file
stopwords_file = set(["english"])

# We leave the stopwords alone here
stopwords_all_words = stopwords.words(stopwords_file)
print(stopwords_all_words)
```

```python
# Creating a function, since we'll need to reuse it for the modern words
def words(word_list):
    result = {}
    for s in word_list:
        # Remove "words" that are caused by splitting contractions.
        if not s in stopwords_all_words: #list(['d','m','s','t','th']):
            if s in result:
                result[s] = result[s] + 1
            else:
                result[s] = 1
    return result

shakespeare_updated_word = words(shakespeare_letter_words)

# Display all shapespeare words without punctuationa and stopwrods
print("\nRemoving stopwords and punctuation the total words in shakespeare are: ", len(shakespeare_updated_word))
print("Shakespeare updated words are:", shakespeare_updated_word)
```

```python
# Sorting the words by their frequency, descending
def sorting_word_counts(word_counts):
    sorted_words = sorted(word_counts, key=lambda item : word_counts[item], reverse=True)
    
    result = []
    for s in sorted_words:
        result.append({'word':s, 'count':word_counts[s]})

    return result

sorted_shakespeare_list = sort_word_counts(shakespeare_updated_word)
shakespeare_top50words = sorted_shakespeare_list[0:50]

shakespeare_top50words_list = [item["word"] for item in shakespeare_top50words]

# Displaying top 50 shakespeare words
print("\nTop 50 Shakespeare words are:",shakespeare_top50_words_list)
```
